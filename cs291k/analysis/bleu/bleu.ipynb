```python
import os
import torch as th
```

```python
path_train = '../tmp/train_beam.txt'
path_gen = '../tmp/gen_beam.txt'
```

```python
def read(path):
    lines = []
    with open(path, 'r') as r:
        for line in r.readlines():
            line = line.strip()
            if len(line) > 0:
                lines.append(line)
    assert len(lines) % 2 == 0
    pairs = [(lines[i * 2], lines[i * 2 + 1]) for i in range(len(lines) // 2)]
    return pairs
```

```python
pairs_train = read(path_train)
pairs_gen = read(path_gen)
```

```python
pairs_train = sorted(pairs_train)
pairs_gen = sorted(pairs_gen)
```

```python
for p_t, p_g in zip(pairs_train, pairs_gen):
    if p_t[1] != p_g[1]:
        print(p_t[0], p_t[1], p_g[0], p_g[1], '\n', sep='\n')
```
