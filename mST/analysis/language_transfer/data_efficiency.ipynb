```python
import os
import random

import numpy as np

from examples.speech_to_text.data_utils import load_df_from_tsv, save_df_to_tsv
```

```python
root = '/mnt/data/siqiouyang/datasets/covost2'
```

```python
lang = 'de'
```

```python
empty = '[EMPTY]'
```

```python
src_df = load_df_from_tsv(os.path.join(root, lang, 'train_st_{}_en.tsv'.format(lang)))
```

```python
ratios = [0.25, 0.5, 0.75]
```

```python
n_sample = len(src_df)
indices = list(range(n_sample))
```

```python
random.shuffle(indices)
```

```python
for idx, ratio in enumerate(ratios):
    st_indices = indices[:int(n_sample * ratio)]

    mask = np.zeros(n_sample, dtype=bool)
    mask[st_indices] = True

    st_df = src_df[mask]
    # os.system('ln -s {} {}'.format(os.path.join(root, lang), os.path.join(root, lang + '-{}'.format(idx))))
    save_df_to_tsv(st_df, os.path.join(root, lang, 'train_st_{}-{}_en.tsv'.format(lang, idx)))

    asr_df = src_df[~mask]
    save_df_to_tsv(asr_df, os.path.join(root, lang, 'train_asr_{}-{}.tsv'.format(lang, idx)))
```
